```python
import pandas as pd
```

#### MERGE MASTER FILE (SCORES_PREP) WITH THE MOVING AVERAGE & TIME DELTA FEATURES

##### Step 1: retrieving and cleaning the files

```python
# reading csv files
data_master = pd.read_csv('../04_datasets/nfl_dataset.csv') # the master file
data1 = pd.read_csv('../04_datasets/nfl_team_ma.csv') # the file with the moving average + the time delta
```

```python
data_master = data_master.drop(columns=['Unnamed: 0'])
data_master.head()
```

```python
data1 = data1.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
data1 = data1.rename(columns = {'index':'game_date'})
data1
```

##### Step 2: merging the files

```python
 # using merge function by setting how='left' to keep in line with the master file
output1 = pd.merge(data_master, data1, 
                   on='idgame', 
                   how='left')
  
# displaying result
output1.head()
```

#### MERGE MASTER DATAFRAME (output1) WITH THE GAME INFO FEATURES

```python
data2 = pd.read_csv('../04_datasets/games_info_features.csv') # the file with the different info on the games (stadium, weather, etc.)
```

```python
data2.head()
```

```python
data2 = data2.drop(columns=['Unnamed: 0'])
```

```python
 # using merge function by setting how='left' to keep in line with the master file
output2 = pd.merge(output1, data2, 
                   on='idgame', 
                   how='left')
  
# displaying result
output2.head()
```

```python
# Creation of the csv
output2.to_csv('nfl_dataset_v1.csv')
```

#### MERGE MASTER DATAFRAME (output2) WITH THE WIN STREAK FEATURE
