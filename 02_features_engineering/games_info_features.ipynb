```python
# import libraries

import os
import pandas as pd
```

```python
os.getcwd()
```

```python
# import file json as dadtaframe
df_gamesinfo_1 = pd.read_json("../01_scraping/json/gamescast.json")
df_gamesinfo_2 = pd.read_json("../01_scraping/json//gamescast_2.json")
df_gamesinfo = pd.concat([df_gamesinfo_1, df_gamesinfo_2], axis =0)
```

```python
df_gamesinfo.head()
```

```python
# describe
df_gamesinfo.describe(include="all")
```

```python
df_gamesinfo.info()
```

```python
df_gamesinfo["attendance"].isnull().sum()
```

```python
# preprocessing attendance

df_gamesinfo["attendance"] = df_gamesinfo["attendance"].apply(lambda x : None if x == None
                                    else int(x.replace("Attendance: ","").replace(",","")))
df_gamesinfo["attendance"].head()
```

```python
# preprocessing capacity
df_gamesinfo["capacity"] = df_gamesinfo["capacity"].apply(lambda x : None if x == None
                                                        else int(x.replace("Capacity: ","").replace(",","")))
df_gamesinfo["capacity"].head()
```

```python
# preprocessing people

df_gamesinfo["people"] = df_gamesinfo["people"].apply(lambda x : None if x == None
                                                    else int(x.replace("%","")))
df_gamesinfo["people"].head()
```

```python
# new feature attendance_info : boolean
df_gamesinfo["attendance_info"] = df_gamesinfo["attendance"].apply(lambda x : 1 if x > 0
                                                                                else 0)
```

```python
df_gamesinfo.describe()
```

```python
df_gamesinfo[df_gamesinfo["attendance_info"]==0].head()
```

```python
df_gamesinfo["date"].dtype
```

```python
# creates new variables with dt

df_gamesinfo['month'] = df_gamesinfo['date'].dt.month
df_gamesinfo['dayofmonth'] = df_gamesinfo['date'].dt.day
df_gamesinfo['dayofweek']=df_gamesinfo["date"].dt.dayofweek
df_gamesinfo['hour'] = df_gamesinfo['date'].dt.hour
df_gamesinfo['minute'] = df_gamesinfo['date'].dt.minute
```
